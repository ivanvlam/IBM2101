```python
import matplotlib.pyplot as plt
import numpy as np
```

### Pregunta 2

```python
def funcion(s):
    if abs(s) <= 1:
        return 1 - abs(s)
    return 0
```

```python
x = np.arange(-2, 2, 0.01)
```

```python
plt.plot(x, tuple(map(funcion, x)), color = 'k')
# plt.title('$\phi\ (s)$')
plt.show()
```

```python
t = 1 / 3310
c1 = 1435
```

```python
def p_p(x, t = t, c1 = c1):
    return funcion(x - c1 * t)

def p_m(x, t = t, c1 = c1):
    return funcion(x + c1 * t)
```

```python
fig = plt.figure()

plt.plot(x, tuple(map(p_p, x)), label = '$p_{+}$')
plt.plot(x, tuple(map(p_m, x)), label = '$p_{-}$')

plt.legend()
plt.show()
```
